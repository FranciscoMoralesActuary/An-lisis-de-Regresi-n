# Arrays-arrange,zeros,ones,linspace,eye

```
#arange genera un arran de n-1 elementos
array=np.arange(5)
array
```

```
array=np.arange(2,10)
array
```

```
#incremento de n en p secuencias
array=np.arange(2,10,2)
array
```

```
#método de zeros para rellear
np.zeros(3)
```

```
#matriz de ceros (renglones, columnas)
np.zeros((3,4))
```

```
#matriz de unos
np.ones((2,5))
```

```
#linspace; matriz de una sola dimensión, de n a p y R elementos
#los incrementos los hace entre el rango (n-p)/(R-1)
#ejemplo para (5,8,50) -> (5-8)/(50-1)=0.0612244
np.linspace(5,8,50)
```

```
np.linspace(3,9,20)
```

```
#eye genera matrices identidades
np.eye(5)
```

```
np.eye(10)
```
