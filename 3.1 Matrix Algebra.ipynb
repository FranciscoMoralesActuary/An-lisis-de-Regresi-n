```python
import numpy as np
from scipy import linalg

A = np.array([[1, 1, 1], [1, 2, 4], [1, 2, 4]])
b = np.array([5, 13, 25]).reshape((3, 1))

p, *_ = linalg.lstsq(A, b)
```

```python
p
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 3, 1000)
y = p[0] + p[1] * x + p[2] * x ** 2

plt.plot(x, y)
plt.plot(1, 5, "ro")
plt.plot(2, 13, "ro")
plt.plot(2, 25, "ro")
```

Pseudo Inverse

```python
import numpy as np
from scipy import linalg

A = np.array([[1, 1, 1], [1, 2, 4], [1, 2, 4]])
b = np.array([5, 13, 25]).reshape((3, 1))

A_pinv = linalg.pinv(A)

p2 = A_pinv @ b
p2
```
