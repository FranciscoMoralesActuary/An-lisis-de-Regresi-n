# Merge DataFrames

```
diccionario={'A':[1,2,3],'B':[4,5,6],'clave':['c1','c2','c3']}
dataframe1=pd.DataFrame(diccionario)
dataframe1
```

```
diccionario2={'A':[11,12,13],'B':[14,15,16],'clave':['c1','c2','c3']}
dataframe2=pd.DataFrame(diccionario2)
dataframe2
```

```
#Unir dos DataFrames (mismos nombres de columnas)
pd.merge(dataframe1,dataframe2, on='clave')
```

```
diccionario3={'C':[11,12,13],'D':[14,15,16],'clave':['c1','c2','c3']}
dataframe3=pd.DataFrame(diccionario3)
dataframe3
```

```
#Unir dos DataFrames (diferentes nombres de columnas)
pd.merge(dataframe1,dataframe3, on='clave')
```
